```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split


import warnings
warnings.filterwarnings("ignore")
```

# SLR :

simple Linear regression :


# Dependent var:  disp
# Independent var:  cyl

```python
A=df[['cyl']]   # Independent variable
B=df[['disp']]  # Dependent
```

#           Train              Test
    A_train.              A_test.
      B_train.              B_test.

```python
# Train_ Test_Split:

A_train,A_test,B_train,B_test=train_test_split(A,B,test_size=0.20,random_state=0)     
```

```python
reg=LinearRegression()
```

```python
# Deployed Simple Linear regression Model :

model=reg.fit(A_train,B_train)
model
```

```python
B_train.shape,B_test.shape                                 
```

```python
A_train.shape,A_test.shape
```

```python
model.intercept_      #values of constant'c'
```

```python
intercept=model.intercept_
intercept
```

```python
model.coef_
```

```python
coef=model.coef_
coef
```

```python
# Checking Model Accuracy 


model.score(A_train,B_train)
```

# Checking Model Accuracy  ‚úÖüòçüòçüéâü•≥= 82%


```python
# Y=mx+c


# Taxable value (‚Çπ)    :  Dependent var.
# Invoice value (‚Çπ)   : Independent  var.


equation='cyl='+str(intercept)+'+disp*'+str(coef)
equation
```

# Prediction :üòéüòé

```python
#      A=df[['cyl']]   # Independent variable
#      B=df[['disp']]  # Dependent



#          Train      Test
    
#  A       A_train.    A_test.

#  B       B_train.    B_test.
```

```python
p_pred=model.predict(A_test)      # test results of Independnet variable
```

```python
p_pred
```

```python
# Visualization  Of result ;


plt.scatter(np.array(A_test),B_test,color='blue')
plt.plot(np.array(A_test),np.array(p_pred),color='red',linewidth=3)
plt.title("Prediction Model Graph Applied on cyl")
plt.xlabel('Invoice Values in future ')
plt.ylabel("Estimated disp in Future");
```
